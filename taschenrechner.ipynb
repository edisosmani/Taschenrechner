## Beispiel 2
### Aufgabenstellung
- Ein weiterer User Input "Welche Rechenoperation soll durchgeführt werden?" --> Speichern Sie diesen Input ebenfalls in einer Variableausgegeben werden und zusätzlich eine Nachricht mit "Nur Zahlen sind erlaubt!"
- Erweitern Sie die Funktionalität mit folgenden Rechenoperationen: Subtraktion, Division und Multiplikation --> Tip: if / elif
- Sollte bei der Auswahl der Rechenoperation keine richtige Auswahl getroffen worden sein soll es zu folgender Ausgabe kommen --> "Keine gültige Rechenoperation"

````python
text = """
=====================================
==         Taschenrechner          ==
=====================================
"""
print(text)
num1 = float(input("Zahl1: "))
num2 = float(input("Zahl2: "))
op = str(input("Rechenzeichen: "))
print("Zahl1: ", num1)
print("Zahl2: ", num2)

if op == "+":
    sum = num1 + num2
    print("Die Summe: ", sum)
elif op == "-":
    diff = num1 - num2
    print("Die Differenz: ", diff)
elif op == "*":
    product = num1 * num2
    print("Die Multiplikation: ", product)
elif op == "/":
    quotient = num1 / num2
    print("Die Division: ", quotient)
else:
    print("Rechenzeichen nicht gefunden!")


````

## Beispiel 3
### Aufgabenstellung
- Bei Eingabe der Zahlen soll, wenn z.B. ein Buchstabe oder ein Zeichen eingegeben wird, das Abstürzen des Programms verhindert werden. Stattdessen soll die Fehlermeldung ausgegeben werden und zusätzlich eine Nachricht mit "Nur Zahlen sind erlaubt!"
- Vor dem Ergebnis soll noch zusätzlich folgendes ausgegeben werden: "Ihre Rechnung: + die zuvor eingegeben Rechnung z.B. 2 * 4"


````python
text = """
=====================================
==         Taschenrechner          ==
=====================================
"""
print(text)

while True:
    try:
        num1 = float(input("Zahl1: "))
        break
    except ValueError:
        print(f"Die Eingabe '{num1}' ist keine gültige Zahl, bitte erneut.")
print("Eingabe akzeptiert:", num1)

while True:
    try:
        num2 = float(input("Zahl2: "))
        break
    except ValueError:
        print(f"Die Eingabe '{num2}' ist keine gültige Zahl")
print("Eingabe akzeptiert:", num2)

op = str(input("Rechenzeichen: "))
print("Zahl1: ", num1)
print("Zahl2: ", num2)
print("Ihre Rechnung: ", num1, op, num2)

if op == "+":
    sum = num1 + num2
    print("Die Summe: ", sum)
elif op == "-":
    diff = num1 - num2
    print("Die Differenz: ", diff)
elif op == "*":
    product = num1 * num2
    print("Die Multiplikation: ", product)
elif op == "/":
    quotient = num1 / num2
    print("Die Division: ", quotient)
else:
    print("Rechenzeichen nicht gefunden!")


````

## Beispiel 4
### Aufgabenstellung
- Definieren Sie eine Funktion mit dem Namen "Taschenrechner" und integrieren Sie Ihren bisherigen Code in diese Funktion
- Rufen Sie die Funktion auf

````python

def taschenrechner():
    text = """
            =====================================
            ==         Taschenrechner          ==
            =====================================
            """
    print(text)

    while True:
        usrinput = str(input("Zahl 1: "))
        try:
            num1 = float(usrinput)
            break
        except ValueError:
            print(f"Die Eingabe '{usrinput}' ist keine gültige Zahl, bitte erneut.")
    print("Eingabe akzeptiert:", num1)
    while True:
        usrinput = str(input("Zahl 2: "))
        try:
            num2 = float(usrinput)
            break
        except ValueError:
            print(f"Die Eingabe '{usrinput}' ist keine gültige Zahl")
    print("Eingabe akzeptiert:", num2)
    op = str(input("Rechenzeichen: "))
    print("Zahl1: ", num1)
    print("Zahl2: ", num2)
    print("Ihre Rechnung: ", num1, op, num2)

    if op == "+":
        sum = num1 + num2
        print("Die Summe: ", sum)
    elif op == "-":
        diff = num1 - num2
        print("Die Differenz: ", diff)
    elif op == "*":
        product = num1 * num2
        print("Die Multiplikation: ", product)
    elif op == "/":
        if num2 == 0:
            print("Division durch null nicht erlaubt")
        else:
            quotient = num1 / num2
            print("Die Division: ", quotient)
    else:
        print("Rechenzeichen nicht gefunden!")


taschenrechner()
````

## Beispiel 5
### Aufgabenstellung
- Verändern Sie den Taschenrechner so, dass beim erfolgreichen Abschluss einer Rechnung, aber auch bei einer Fehlermeldung die Funktion neu aufgerufen wird und somit in einer Schleife läuft.
- Integrieren Sie einen Weg zum Beenden der Schleife (z.B. mit der Eingabe des Buchstabens „q“).

````python

def taschenrechner():
    text = """
            =====================================
            ==         Taschenrechner          ==
            =====================================
            """
    print(text)

    while True:
        usrinput = str(input("Zahl 1 oder 'q' zum Beenden: "))
        if usrinput.lower() == "q":
            print("Taschenrechner wird beendet.")
            break
        try:
            num1 = float(usrinput)
        except ValueError:
            print(f"Die Eingabe '{usrinput}' ist keine gültige Zahl, bitte erneut.")
            continue
        print("Eingabe akzeptiert:", num1)

        usrinput = str(input("Zahl 2 oder 'q' zum Beenden: "))
        if usrinput == "q":
            print("Taschenrechner wird beendet.")
            break
        try:
            num2 = float(usrinput)
        except ValueError:
            print(f"Die Eingabe '{usrinput}' ist keine gültige Zahl")
            continue
        print("Eingabe akzeptiert:", num2)


        op = str(input("Rechenzeichen (+, -, *, / oder 'q' zum Beenden"))
        if op.lower() == "q":
            print("Taschenrechner wird beendet.")
            break

        print("Zahl1: ", num1)
        print("Zahl2: ", num2)
        print("Ihre Rechnung: ", num1, op, num2)

        if op == "+":
            sum = num1 + num2
            print("Die Summe: ", sum)
        elif op == "-":
            diff = num1 - num2
            print("Die Differenz: ", diff)
        elif op == "*":
            product = num1 * num2
            print("Die Multiplikation: ", product)
        elif op == "/":
            if num2 == 0:
                print("Division durch null nicht erlaubt")
            else:
                quotient = num1 / num2
                print("Die Division: ", quotient)
        else:
            print("Rechenzeichen nicht gefunden!")

taschenrechner()
````
