```python
text = """
=====================================
==         Taschenrechner          ==
=====================================
"""
print(text)
num1 = float(input("Zahl1: "))
num2 = float(input("Zahl2: "))
op = str(input("Rechenzeichen: "))
print("Zahl1: ", num1)
print("Zahl2: ", num2)

if op == "+":
    sum = num1 + num2
    print("Die Summe: ", sum)
elif op == "-":
    diff = num1 - num2
    print("Die Differenz: ", diff)
elif op == "*":
    product = num1 * num2
    print("Die Multiplikation: ", product)
elif op == "/":
    quotient = num1 / num2
    print("Die Division: ", quotient)
else:
    print("Rechenzeichen nicht gefunden!")


```
