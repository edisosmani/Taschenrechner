## Beispiel 2

```python
text = """
=====================================
==         Taschenrechner          ==
=====================================
"""
print(text)
zahl1 = float(input("Zahl1: "))
zahl2 = float(input("Zahl2: "))
rechenzeichen = str(input("Rechenzeichen: "))
print("Zahl1: ", zahl1)
print("Zahl2: ", zahl2)

if rechenzeichen == "+":
    sum = zahl1 + zahl2
    print("Die Summe: ", sum)
elif rechenzeichen == "-":
    diff = zahl1 - zahl2
    print("Die Differenz: ", diff)
elif rechenzeichen == "*":
    product = zahl1 * zahl2
    print("Die Multiplikation: ", product)
elif rechenzeichen == "/":
    quotient = zahl1 / zahl2
    print("Die Division: ", quotient)
else:
    print("Rechenzeichen nicht gefunden!")


```
